```python
import os
from PIL import Image, ImageFilter
import numpy as np
import pandas as pd
from skimage import color, filters, transform
import matplotlib.pyplot as plt
```

```python
image_folder = '/content/drive/MyDrive/Módulos/Módulo 10/pond_pre_processamento/img/'
image_files = os.listdir(image_folder)[:10]  # Load only the first 10 images
images = [Image.open(os.path.join(image_folder, file)) for file in image_files]
```

```python
def resize_image(image, target_size=(128, 128)):
    return image.resize(target_size)

def convert_to_grayscale(image):
    return image.convert('L')

def apply_gaussian_blur(image):
    return image.filter(ImageFilter.GaussianBlur(radius=2))
```

```python
# Expected results (you would replace these with actual expected values or examples)
expected_sizes = [(128, 128)] * 10
expected_modes = ['L'] * 10

# Processed images
processed_images = [resize_image(img) for img in images]
grayscale_images = [convert_to_grayscale(img) for img in processed_images]
blurred_images = [apply_gaussian_blur(img) for img in grayscale_images]

# Test functions
def test_image_sizes(images, expected_sizes):
    actual_sizes = [img.size for img in images]
    assert actual_sizes == expected_sizes, f"Sizes do not match: {actual_sizes} != {expected_sizes}"

def test_image_modes(images, expected_modes):
    actual_modes = [img.mode for img in images]
    assert actual_modes == expected_modes, f"Modes do not match: {actual_modes} != {expected_modes}"

# Run tests
test_image_sizes(processed_images, expected_sizes)
test_image_modes(grayscale_images, expected_modes)
```

```python
# Convert images to numpy arrays
image_arrays = [np.array(img) for img in blurred_images]

# Create DataFrame
df = pd.DataFrame({
    'Image Name': image_files,
    'Processed Image': image_arrays
})
df.head()
```
